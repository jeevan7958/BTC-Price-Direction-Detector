```python
import numpy as np
import pandas as pd
import shap
import tensorflow as tf
from tensorflow.keras import layers, models, callbacks
import joblib
import matplotlib.pyplot as plt

# --- STEP 1: LOAD DATA ---
X = np.load('../data/processed/X.npy')
y = np.load('../data/processed/y.npy')

# Split Data (80/20)
split_index = int(len(X) * 0.8)
X_train, X_test = X[:split_index], X[split_index:]
y_train, y_test = y[:split_index], y[split_index:]

# --- STEP 2: RE-BUILD & RE-TRAIN MODEL ---
# (Define 'model' so SHAP has something to explain)

def transformer_encoder(inputs, head_size, num_heads, ff_dim, dropout=0):
    x = layers.MultiHeadAttention(
        key_dim=head_size, num_heads=num_heads, dropout=dropout
    )(inputs, inputs)
    x = layers.Dropout(dropout)(x)
    x = layers.LayerNormalization(epsilon=1e-6)(x)
    res = x + inputs
    x = layers.Conv1D(filters=ff_dim, kernel_size=1, activation="relu")(res)
    x = layers.Dropout(dropout)(x)
    x = layers.Conv1D(filters=inputs.shape[-1], kernel_size=1)(x)
    x = layers.LayerNormalization(epsilon=1e-6)(x)
    return x + res

def build_model(input_shape):
    inputs = layers.Input(shape=input_shape)
    x = inputs
    x = transformer_encoder(x, head_size=128, num_heads=2, ff_dim=4, dropout=0.1)
    x = transformer_encoder(x, head_size=128, num_heads=2, ff_dim=4, dropout=0.1)
    x = layers.GlobalAveragePooling1D(data_format="channels_first")(x)
    x = layers.Dense(64, activation="relu")(x)
    x = layers.Dropout(0.1)(x)
    outputs = layers.Dense(1)(x)
    return models.Model(inputs, outputs)

# Build & Train (Quickly)
print("Re-training model for interpretation...")
model = build_model(input_shape=(X_train.shape[1], X_train.shape[2]))
model.compile(loss="mean_squared_error", optimizer=tf.keras.optimizers.Adam(learning_rate=0.0001))
model.fit(X_train, y_train, epochs=10, batch_size=32, verbose=0) # Silent training
print("Model ready.")

# --- STEP 3: SHAP EXPLANATION ---

# Take a small sample (SHAP is slow!)
explainer_data = X_test[:50] 
X_test_2d = explainer_data.reshape(explainer_data.shape[0], -1)

# Wrapper to reshape 2D SHAP input back to 3D for the model
def model_predict_wrapper(data_2d):
    data_3d = data_2d.reshape(-1, 14, 2)
    return model.predict(data_3d).flatten()

# Create Explainer
explainer = shap.Explainer(model_predict_wrapper, X_test_2d)

print("Calculating SHAP values (this might take a minute)...")
shap_values = explainer(X_test_2d)

# --- STEP 4: VISUALIZE ---
# Feature 0 = Day1_Return, Feature 1 = Day1_Volume, ...
feature_names = []
for i in range(14):
    feature_names.append(f"Day{i+1}_Return")
    feature_names.append(f"Day{i+1}_Vol")

shap_values.feature_names = feature_names

plt.figure(figsize=(10, 6))
plt.title("What drives the prediction?")
shap.summary_plot(shap_values, X_test_2d, feature_names=feature_names)

# Save the model to the hard drive
model.save('../models/transformer_model.keras')
print("Model saved! Refresh your Streamlit app.")
```
